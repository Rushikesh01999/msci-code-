## Handeling year 2020 & 2021 as there's no data for those 2 years

### Removing Duplicate column of 1996

```python
carbon_ems=carbon_ems.drop('1996.1',axis=1)
```

## Handeling NAN Values in countries, Either to drop thode countries or considered 0 

```python
 carbon_ems['None']=carbon_ems.isnull().sum(axis=1)
```

```python
carbon_ems[carbon_ems['None']>25]=carbon_ems[carbon_ems['None']>25].fillna(0)
```

```python
carbon_ems['mean'] = carbon_ems.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code','None','1994','1995','1996'],axis=1).mean(axis=1)
```

```python
carbon_ems['max'] = carbon_ems.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code','None','1994','1995','1996'],axis=1).max(axis=1)
```

### Few Countri's region is not given

```python
merged_df=carbon_ems.merge(Region_details,on='Country Code',how='left')
```

```python
merged_df[merged_df['IncomeGroup'].isna()]
```

```python
merged_df['co2_emission'] = merged_df.drop(['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code','None','mean','max','Region','IncomeGroup','SpecialNotes','TableName'],axis=1).sum(axis=1)
```

```python
sorted_df=merged_df.groupby(['Region','IncomeGroup'])['co2_emission'].sum().reset_index().sort_values(['Region','IncomeGroup'])
```
